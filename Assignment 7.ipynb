```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

r= 102317116
np.random.seed(r)
sales_data = np.random.randint(1000,5001,size = (12,4))

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
categories = ["Home","Clothing","Electronics","Sports"]
df = pd.DataFrame(sales_data,columns = categories,index = months)

df
```

```python
# Display first 5 rows
print(df.head())
print()

# Summary statistics
print(df.describe())
print()

total_sales_category = df.sum()
print(total_sales_category)
print()

df["Total Sales"] = df.sum(axis=1)
df
print()

df["Growth Rate"] = df["Total Sales"].pct_change() * 100
df
print()

if r % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.90  
else:
    df["Clothing"] = df["Clothing"] * 0.85  

df
print()
```

```python
# Line plot for sales trends
plt.figure(figsize = (10,6))
for category in categories:
    plt.plot(df.index,df[category],marker = "o",label = category)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.show()

# Box plot for sales distribution
plt.figure(figsize = (8,6))
sns.boxplot(data = df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_values = np.abs(array)
print(abs_values)
percentiles = np.percentile(array, [25, 50, 75], axis=None)
print(percentiles)
mean_val = np.mean(array)
median_val = np.median(array)
std_dev = np.std(array)
print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_dev)
```

```python
b = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_values = np.floor(b)
ceiling_values = np.ceil(b)
truncated_values = np.trunc(b)
rounded_values = np.round(b)
print(floor_values)
print(ceiling_values)
print(truncated_values)
print(rounded_values)
```
